```python
import random

def setWeights(noOfInputs:int, noOfNodes:int, noOfLayers:int)->list:
    totalWeights = []
    for eachLayer in range(0, noOfLayers, 1):
        layer = []
        for eachNode in range(0, noOfNodes, 1):
            node = []
            for eachWeight in range(0, noOfInputs, 1):
                weight = random.random()
                node.append(weight)
            layer.append(node)
        totalWeights.append(layer)
    return totalWeights
```

```python
def setBiases(noOfNodes:int, noOfLayers:int)->list:
    totalBiases = []
    for eachLayer in range(0, noOfLayers, 1):
        layer = []
        for eachNode in range(0, noOfNodes, 1):
            weight = random.random()
            layer.append(weight)
        totalBiases.append(layer)
    return totalBiases
```

```python
def matrixProduct(inputs:list, weights:list)->float:
    productSum = 0
    for i in range(0, len(inputs), 1):
        product = weights[i] * inputs[i]
        productSum = productSum + product
    return productSum
```
