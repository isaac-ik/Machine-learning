```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Salary_data = pd.read_csv('../Position_Salaries.csv')
```

```python
Salary_data
```

```python
# Split data into feature and target vector
X = Salary_data.iloc[:, 1:-1].values
y = Salary_data.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
# Perform Decison tree regression
from sklearn.tree import DecisionTreeRegressor
treeRegressor = DecisionTreeRegressor()
treeRegressor.fit(X,y)
```

```python
# Predict a Salary for a level
treeRegressor.predict([[6.5]])
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, treeRegressor.predict(X), color='blue')
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid), 1)
```

```python
plt.scatter(X,y, color='red')
plt.plot(X_grid, treeRegressor.predict(X_grid), color='green')
```

```python
# Visualize the Decision Tree
from sklearn.tree import plot_tree

plt.figure(figsize=(20, 10))
plot_tree(treeRegressor, feature_names=['Level'], filled=True)
plt.show()
```
