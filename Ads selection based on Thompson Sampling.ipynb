```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ads_data = pd.read_csv('../Ads_CTR_Optimisation.csv')
```

```python
import random
N = len(ads_data)
d = len(ads_data.columns)

ads_selected = [] # will contain all ads selected at each round
times_selected = [0] * d
times_notSelected = [0] * d
total_reward = 0
for n in range(0, N): # iterate through all round
    ad = 0 # ad selected for the round
    max_random = 0
    for i in range(0, d): # iterate through all rounds in each round
        j = random.betavariate(times_selected[i] + 1, times_notSelected[i] + 1)
        if j > max_random:
            ad = i
            max_random = j
    ads_selected.append(ad)
    reward = ads_data.values[n, ad]
    if reward == 1:
        times_selected[ad] = times_selected[ad] + 1
    else:
        times_notSelected[ad] = times_notSelected[ad] + 1
    total_reward = total_reward + reward
        
```

```python
plt.hist(ads_selected)
plt.title('Histogram of ads selection')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
